```python
def is_game_complete(board,player1_symbol):
    complete = False
    #check for horizontal chains
    for k in board.keys():
        if board[k][0] == board[k][1] == board[k][2]!=0:
            if (board[k][0]=='X') != (player1_symbol == 'X'):
                print('Player2 wins')
            else:
                print('Player1 wins')
            complete = True
            break
                
    #check for vertical chains
    for i in range(3):
        if board['1'][i] == board['2'][i] == board['3'][i] !=0:
            if (board['1'][i]=='X')!= (player1_symbol == 'X'):
                print('Player2 wins')
            else:
                print('Player1 wins')
            complete = True
            break
    
    #check for diagonals
    if board['1'][0] == board['2'][1] == board['3'][2] != 0:
        if (board['1'][0]=='X') != (player1_symbol == 'X'):
            print('Player2 wins')
        else:
            print('Player1 wins')
        complete = True
    
    if board['1'][2] == board['2'][1] == board['3'][0] != 0:
        if (board['1'][2]=='X') != (player1_symbol == 'X'):
            print('Player2 wins')
        else:
            print('Player1 wins')
        complete = True
        
    #check for stalemate
    stalemate = True
    for k in board.keys():
        for i in range(3):
            if board[k][i]==0:
                stalemate = False
    
    if not complete:
        if stalemate:
            print ("it's a draw!")
    
    return complete or stalemate
```

```python
def display_board(board):
    for k,v in board.items():
        print('   |     |')
        for i in range(3):
            if board[k][i]=='O': 
                print('O ', end = " ")
                if i!=2:
                    print('| ', end=" ")
            elif board[k][i]=='X':
                print ('X ', end = " ")
                if i!=2:
                    print('| ', end=" ")
            else:
                print('  ', end = " ")
                if i!=2:
                    print('| ', end=" ")
        print('\n   |     |')
        if k !='3':
            print("-------------")
        
```

```python
def get_input(board, player1_symbol, player2_symbol, player1_turn):
    while True:
        temp = input('enter player1 input:' if player1_turn else 'enter player2 input:')
        if temp == 'q' or temp == 'Q':
            if board['1'][0] == 0:
                board['1'][0] = player1_symbol if player1_turn else player2_symbol
                break
            else:
                print("slot already taken, invalid input")
        elif temp == 'w' or temp == 'W':
            if board['1'][1] == 0:
                board['1'][1] = player1_symbol if player1_turn else player2_symbol 
                break
            else:
                print("slot already taken, invalid input")
        elif temp == 'e' or temp == 'E':
            if board['1'][2] == 0:
                board['1'][2] = player1_symbol if player1_turn else player2_symbol
                break
            else:
                print("slot already taken, invalid input")
        elif temp == 'a' or temp == 'A':
            if board['2'][0] == 0:
                board['2'][0] = player1_symbol if player1_turn else player2_symbol
                break
            else:
                print("slot already taken, invalid input")
        elif temp == 's' or temp == 'S':
            if board['2'][1] == 0:
                board['2'][1] = player1_symbol if player1_turn else player2_symbol
                break
            else:
                print("slot already taken, invalid input")
        elif temp == 'd' or temp == 'D':
            if board['2'][2] == 0:
                board['2'][2] = player1_symbol if player1_turn else player2_symbol
                break
            else:
                print("slot already taken, invalid input")
        elif temp == 'z' or temp == 'Z':
            if board['3'][0] == 0:
                board['3'][0] = player1_symbol if player1_turn else player2_symbol
                break
            else:
                print("slot already taken, invalid input")
        elif temp == 'x' or temp == 'X':
            if board['3'][1] == 0:
                board['3'][1] = player1_symbol if player1_turn else player2_symbol
                break
            else:
                print("slot already taken, invalid input")
        elif temp == 'c' or temp == 'C':
            if board['3'][2] == 0:
                board['3'][2] = player1_symbol if player1_turn else player2_symbol
                break
            else:
                print("slot already taken, invalid input")
        else:
            print("invalid input")
```

```python
from IPython.display import clear_output

play_game=True
while play_game:
    board = {'1':[0,0,0],'2':[0,0,0],'3':[0,0,0]}
    while True:
        player1_symbol = input('Does player1 wants to be X or O ? ').upper()
        if player1_symbol == 'X' or player1_symbol=='x':
            player2_symbol = 'O'
            break
        elif player1_symbol == 'O' or player1_symbol=='o':
            player2_symbol = 'X'
            break
        else:
            print("invalid input")
    player1_turn = True
    while not is_game_complete(board,player1_symbol):
        clear_output()
        display_board(board)
        get_input(board, player1_symbol, player2_symbol, player1_turn)
        player1_turn = not player1_turn
    else:
        clear_output()
        display_board(board)
        is_game_complete(board,player1_symbol)
    
    while True:
        ans = input("Do you want to play another game? Enter yes or no: ")
        if ans.lower() == 'yes':
            play_game = True
            break
        elif ans.lower() == 'no':
            play_game = False
            break
        else:
            print("invalid input")
else:
    print("Goodbye!")
    
```
